```python
import pandas as pd
```

```python
# loading the product and translation product datasets
products = pd.read_csv(r"C:\Users\Amarjeet\Downloads\Revenue_Leagkage_Analysis\raw_data\products.csv")
translation_prod = pd.read_csv(r"C:\Users\Amarjeet\Downloads\Revenue_Leagkage_Analysis\raw_data\product_translation.csv")
```

```python
products.head()
```

```python
translation_prod.head()
```

```python
# Checking structure, summary, and missing values in products
print(products.shape)
print(products.info())
print(products.describe())
print(products.isnull().sum())
```

```python
# Checking structure, summary, and missing values in translation table
print(translation_prod.shape)
print(translation_prod.info())
print(translation_prod.describe())
print(translation_prod.isnull().sum())
```

```python
# merging product with translation prod table for english translation of product categories

products_merged = pd.merge(products, translation_prod, on = 'product_category_name', how = 'left')
products_merged.head()
```

```python
# renaming the column name and dropping the name column which was in different language after merging
products_merged = products_merged.rename(
    columns = {'product_category_name_english' : 'product_category'})

products_merged.drop(columns = ['product_category_name'], inplace = True)

products_merged.head()
```

```python
# keeping only needed columns for analysis
products = products_merged[
["product_id", "product_category"]]
products.head()
```

```python
# filling the missing product categories with 'unknown'
products.loc[:,'product_category'] = products['product_category'].fillna('Unknown')
```

```python
# checking for any null values after cleaning
print(products.isnull().sum())
print(products.head(10))
```

```python
# checking for any duplicate raws
products['product_id'].duplicated().sum()
```

```python
# removing any extra spaces and standardizing text
products.loc[:, 'product_category'] = products['product_category'].str.strip().str.lower()
```

```python
# saving the file
products_cleaned = products.to_csv(r"C:\Users\Amarjeet\Downloads\Revenue_Leagkage_Analysis\Cleaned_data\products_cleaned.csv", index=False)
```

```python
# loading order item dataset
order_items = pd.read_csv(r'C:\Users\Amarjeet\Downloads\Revenue_Leagkage_Analysis\raw_data\order_items.csv')
order_items.head()
```

```python
# Checking structure, summary, and missing values in order items
print(order_items.shape)
print(order_items.info())
print(order_items.describe())
print(order_items.isnull().sum())
```

```python
# checking for duplicates
order_items.duplicated().sum()
```

```python
# changing the shipping dae to datetime format
order_items['shipping_limit_date'] = pd.to_datetime(order_items['shipping_limit_date'])
```

```python
# creating shipping ratio column for analysis
order_items['shipping_ratio'] = order_items['freight_value']/order_items['price']
```

```python
# summary for shipping ratio
print(order_items['shipping_ratio'].describe())
order_items.head()
```

```python
# finding order items where shipping ratio is high
order_items[order_items['shipping_ratio']>1].shape
```

```python
# (4124/112650)*100 = 3.6(approx) Approximately 3.6% of orders have shipping costs exceeding product price
```

```python
# creating column for high shipping order items
order_items['high_shipping'] = order_items['shipping_ratio'] > 1
```

```python
order_items['high_shipping'].value_counts()
```

```python
# loading the orders dataset
orders = pd.read_csv(r'C:\Users\Amarjeet\Downloads\Revenue_Leagkage_Analysis\raw_data\orders.csv')
```

```python
# Checking structure, summary, and missing values in orders
print(orders.shape)
print(orders.info())
print(orders.describe())
print(orders.isnull().sum())
print(orders.duplicated().sum())
```

```python
orders.head()
```

```python
# changing the columns type to datetime format
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_approved_at'] = pd.to_datetime(orders['order_approved_at'])
orders['order_delivered_carrier_date'] = pd.to_datetime(orders['order_delivered_carrier_date'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])
```

```python
orders.info()
```

```python
# creating delivery days columns
orders['delivery_days'] = (orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']).dt.days
```

```python
# creating column for late deliveries
orders['delivery_delay'] = (orders['order_delivered_customer_date'] - orders['order_estimated_delivery_date']).dt.days
```

```python
# creating true false column for analysis
orders['is_delivery_late'] = orders['delivery_delay']>0
```

```python
# checking for uniques values
orders['order_status'].value_counts()
```

```python
# Checking summary
print(orders['delivery_days'].describe())
print(orders['delivery_days'].describe())
```

```python
# saving the file
orders_cleaned = orders.to_csv(r"C:\Users\Amarjeet\Downloads\Revenue_Leagkage_Analysis\Cleaned_data\orders_cleaned.csv", index = False)
```
